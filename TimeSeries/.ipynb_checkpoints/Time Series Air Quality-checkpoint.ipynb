```python
air['CO(GT)']=air['CO(GT)'].str.replace(',', '.').astype(float)
air['C6H6(GT)']=air['C6H6(GT)'].str.replace(',', '.').astype(float)
air['T']=air['T'].str.replace(',', '.').astype(float)
air['RH']=air['RH'].str.replace(',', '.').astype(float)
air['AH']=air['AH'].str.replace(',', '.').astype(float)
```

```python
air.info()
```

```python
#### Preparing the data. Removing non existing columns and Nans.
```

```python
air.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1, inplace=True)
```

```python
air.shape
```

```python
air.dropna(inplace=True)
```

```python
air.shape
```

```python
air = air[(air['AH']!=-200)] #It would be used those rows with no missing values for AH 
                             # A missing value has -200 value
```

```python
#air.replace(-200,np.nan, inplace=True)
#air.dropna(inplace=True)
air.shape
```

```python
#### EDA
```

```python
air.head()
```

```python
#Transforming European format dd/mm/yyyy to American mm/dd/yyyy
day = air['Date'].str[:2]
month = air['Date'].str[3:5]
year = air['Date'].str[6:]
```

```python
air['Date'] = year +'/'+ month +'/'+ day
```

```python
air['Date'] = pd.to_datetime(air['Date'])
```

```python
air.head()
```

```python
print("The minimum date is", air['Date'].min(), "and the maximum turns out to be", air['Date'].max())
```

```python
print("The total time of data would be",air['Date'].max()-air['Date'].min())
```

```python
print("It is possible to see that it exists different values for each day, therefore the average would be performed for each day")
```

```python
air = air.set_index('Date')
```

```python
air
```

```python
d = air['AH'].resample('D').mean()
m = air['AH'].resample('M').mean()
y = air['AH'].resample('Y').mean()
```

```python
d.head()
```

```python
d.index
```

```python
m.head()
```

```python
y.head()
```

```python
d.isnull().sum()
```

```python
d=d[d.isnull()==False]
```

```python
d.plot(figsize=(15, 6))
m.plot(figsize=(15, 6))
plt.show()
```

```python
#Since the model would be studied by days, the freq is set to 30 (aproximately the days per month)
decomposition_days = sm.tsa.seasonal_decompose(d, model='additive', freq=30)
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 8

fig = decomposition_days.plot()
plt.show()
```
